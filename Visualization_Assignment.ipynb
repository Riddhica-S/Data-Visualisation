```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white",color_codes=True)
```

```python
data = pd.read_csv("train.csv")
```

```python
data.plot(kind="scatter",x="PassengerId",y="Age",color="purple")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
x=data['Sex']
y=data['Pclass']
z=data['Age']
ax.scatter(x,y,z,color="purple")
plt.show()
```

```python
sns.lineplot(x="SibSp",y="Survived",data=data,size=20,color="purple")
plt.show()
```

```python
sns.histplot(x="Age",data=data,color="purple")
plt.show()
```

```python
plt.boxplot(data['Pclass'].fillna(0.0))
```

```python
plt.bar(data['Sex'],data['Parch'])
plt.show()
```
